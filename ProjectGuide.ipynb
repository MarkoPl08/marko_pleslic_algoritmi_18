<h2>Task 1. Implement Hill-Climber Algorithm </h2> which is capable of solving N-dimensional problem. In the code block you can find example objective functions. Feel free to use any multi-dimensional objective function. The idea is that your implementation works the same for 1-dim and N-dim objective functions.

```python
import numpy as np


def objective1(x):
    a = 1
    b = 100
    return (a - x[0]) ** 2 + b * (x[1] - x[0] ** 2) ** 2


def objective2(x):
    sum_part = sum(x_i ** 2 for x_i in x) / 4000
    product_part = np.prod(np.cos(x / np.sqrt(np.arange(1, len(x) + 1))))
    return 1 + sum_part - product_part


def hill_climber(objective_func, bounds, dimensions, max_iterations=1000, step_size=0.1):
    current = np.random.uniform(bounds[0], bounds[1], dimensions)
    current_value = objective_func(current)

    for iteration in range(max_iterations):
        neighbor = current + np.random.normal(0, step_size, dimensions)
        neighbor = np.clip(neighbor, bounds[0], bounds[1])
        neighbor_value = objective_func(neighbor)

        if neighbor_value < current_value:
            current = neighbor
            current_value = neighbor_value

    return current, current_value


x = np.array([0.2, 1.0, 23.3, 1.2, 2.5])
value = objective2(x)
print(f'objective2 function value at {x}: {value}')

x = np.array([1.2, 23.2])
value = objective1(x)
print(f'objective1 function value at {x}: {value}')

result1 = hill_climber(objective1, (-5, 5), 2)
print(f'Hill Climber on Rosenbrock: position={result1[0]}, value={result1[1]}')

result2 = hill_climber(objective2, (-5, 5), 1)
print(f'Hill Climber on Griewank: position={result2[0]}, value={result2[1]}')
```

<h2> Task 3. Implement Simulated Annealing </h2> with following options for "cooling" the temperature:

*   Linear Cooling Schedule: Tk = T0 − α⋅k, where T0 is initial temperature, α is a constant and k is iteration number.
*   Logarithmic Cooling Schedule: Tk = (T0)/(log(k+1)), where T0 is initial temperature and k is iteration number
*   Exponential Cooling Schedule: Tk = T0 ⋅ B^k, where T0 is initial temperature, k is iteration number and 0 < β < 1 is cooling rate
*   Adaptive Cooling Schedule: An adaptive temperature schedule adjusts the cooling rate based on the algorithm's performance, such as the acceptance rate of new solutions. The idea is to slow down the cooling process if the algorithm is still finding better solutions and to speed it up if the algorithm is stagnating
*   Custom cooling schedule: A custom temperature schedule can be designed to fit specific problem characteristics or based on empirical observations. Get creative in this one. It does not have to work better, but have an explanation of your idea.


```python
import numpy as np
import matplotlib.pyplot as plt


def simulated_annealing(objective_func, bounds, dimensions, max_iterations=1000,
                        initial_temp=100, cooling_schedule='exponential',
                        alpha=0.1, beta=0.99, step_size=0.1):
    current = np.random.uniform(bounds[0], bounds[1], dimensions)
    current_value = objective_func(current)
    best = current.copy()
    best_value = current_value

    acceptance_count = 0
    check_interval = 50

    for k in range(max_iterations):
        neighbor = current + np.random.normal(0, step_size, dimensions)
        neighbor = np.clip(neighbor, bounds[0], bounds[1])
        neighbor_value = objective_func(neighbor)

        if cooling_schedule == 'linear':
            temperature = max(0.01, initial_temp - alpha * k)
        elif cooling_schedule == 'logarithmic':
            temperature = initial_temp / np.log(k + 2)
        elif cooling_schedule == 'exponential':
            temperature = initial_temp * (beta ** k)
        elif cooling_schedule == 'adaptive':
            if k > 0 and k % check_interval == 0:
                acceptance_rate = acceptance_count / check_interval
                if acceptance_rate > 0.3:
                    beta = 0.995
                elif acceptance_rate < 0.1:
                    beta = 0.95
                else:
                    beta = 0.99
                acceptance_count = 0
            temperature = initial_temp * (beta ** k)
        elif cooling_schedule == 'custom':
            wave = 0.3 * np.sin(2 * np.pi * k / 100)
            temperature = initial_temp * (beta ** k) * (1 + wave)

        if neighbor_value < current_value:
            current = neighbor
            current_value = neighbor_value
            if cooling_schedule == 'adaptive':
                acceptance_count += 1
        else:
            delta = neighbor_value - current_value
            probability = np.exp(-delta / temperature) if temperature > 0 else 0
            if np.random.random() < probability:
                current = neighbor
                current_value = neighbor_value
                if cooling_schedule == 'adaptive':
                    acceptance_count += 1

        if current_value < best_value:
            best = current.copy()
            best_value = current_value

    return best, best_value


schedules = ['linear', 'logarithmic', 'exponential', 'adaptive', 'custom']
results = {}

print("Simulated Annealing Results:")
print("=" * 60)

for schedule in schedules:
    solution, value = simulated_annealing(objective2, (-5, 5), 5,
                                          max_iterations=2000,
                                          cooling_schedule=schedule)
    results[schedule] = (solution, value)
    print(f"{schedule.upper():12} | value={value:.6f} | position={solution}")

print("=" * 60)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
iterations = np.arange(1000)
for schedule in ['linear', 'logarithmic', 'exponential']:
    temps = []
    for k in iterations:
        if schedule == 'linear':
            temp = max(0.01, 100 - 0.1 * k)
        elif schedule == 'logarithmic':
            temp = 100 / np.log(k + 2)
        elif schedule == 'exponential':
            temp = 100 * (0.99 ** k)
        temps.append(temp)
    plt.plot(iterations, temps, label=schedule.title())
plt.xlabel('Iterations')
plt.ylabel('Temperature')
plt.title('Cooling Schedules Comparison')
plt.legend()
plt.yscale('log')

plt.subplot(2, 2, 2)
schedule_names = list(results.keys())
values = [results[s][1] for s in schedule_names]
colors = ['red', 'blue', 'green', 'orange', 'purple']
bars = plt.bar(schedule_names, values, color=colors, alpha=0.7)
plt.xlabel('Cooling Schedule')
plt.ylabel('Final Objective Value')
plt.title('Performance Comparison')
plt.xticks(rotation=45)

for bar, value in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f'{value:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time


def rastrigin(x):
    A = 10
    n = len(x)
    return A * n + sum(x_i ** 2 - A * np.cos(2 * np.pi * x_i) for x_i in x)


def simulated_annealing_metrics(objective_func, bounds, dimensions, max_iterations=1000,
                                initial_temp=100, cooling_schedule='exponential',
                                alpha=0.1, beta=0.99, step_size=0.1):
    current = np.random.uniform(bounds[0], bounds[1], dimensions)
    current_value = objective_func(current)
    best = current.copy()
    best_value = current_value

    worse_accepted = 0
    convergence_iter = max_iterations
    check_interval = 50
    acceptance_count = 0

    for k in range(max_iterations):
        neighbor = current + np.random.normal(0, step_size, dimensions)
        neighbor = np.clip(neighbor, bounds[0], bounds[1])
        neighbor_value = objective_func(neighbor)

        if cooling_schedule == 'linear':
            temperature = max(0.01, initial_temp - alpha * k)
        elif cooling_schedule == 'logarithmic':
            temperature = initial_temp / np.log(k + 2)
        elif cooling_schedule == 'exponential':
            temperature = initial_temp * (beta ** k)
        elif cooling_schedule == 'adaptive':
            if k > 0 and k % check_interval == 0:
                acceptance_rate = acceptance_count / check_interval
                if acceptance_rate > 0.3:
                    beta = 0.995
                elif acceptance_rate < 0.1:
                    beta = 0.95
                else:
                    beta = 0.99
                acceptance_count = 0
            temperature = initial_temp * (beta ** k)
        elif cooling_schedule == 'custom':
            wave = 0.3 * np.sin(2 * np.pi * k / 100)
            temperature = initial_temp * (beta ** k) * (1 + wave)

        if neighbor_value < current_value:
            current = neighbor
            current_value = neighbor_value
            if cooling_schedule == 'adaptive':
                acceptance_count += 1
        else:
            delta = neighbor_value - current_value
            probability = np.exp(-delta / temperature) if temperature > 0 else 0
            if np.random.random() < probability:
                current = neighbor
                current_value = neighbor_value
                worse_accepted += 1
                if cooling_schedule == 'adaptive':
                    acceptance_count += 1

        if current_value < best_value:
            best = current.copy()
            best_value = current_value
            if best_value < 0.1 and convergence_iter == max_iterations:
                convergence_iter = k

    exploration_rate = worse_accepted / max_iterations
    return best, best_value, exploration_rate, convergence_iter


schedules = ['linear', 'logarithmic', 'exponential', 'adaptive', 'custom']
num_trials = 30
results = {schedule: {'quality': [], 'convergence': [], 'exploration': []} for schedule in schedules}

for schedule in schedules:
    for trial in range(num_trials):
        start_time = time.time()
        _, quality, exploration, convergence = simulated_annealing_metrics(
            rastrigin, (-5.12, 5.12), 2, max_iterations=1000, cooling_schedule=schedule
        )
        execution_time = time.time() - start_time

        results[schedule]['quality'].append(quality)
        results[schedule]['convergence'].append(convergence)
        results[schedule]['exploration'].append(exploration)

print("Schedule | Mean Quality | Convergence Speed | Exploration Rate")
print("-" * 65)
for schedule in schedules:
    mean_quality = np.mean(results[schedule]['quality'])
    mean_convergence = np.mean(results[schedule]['convergence'])
    mean_exploration = np.mean(results[schedule]['exploration'])
    print(f"{schedule:12} | {mean_quality:11.3f} | {mean_convergence:15.1f} | {mean_exploration:12.3f}")

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
quality_data = [results[s]['quality'] for s in schedules]
plt.boxplot(quality_data, tick_labels=schedules)
plt.ylabel('Final Objective Value')
plt.title('Quality of Solutions')
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
convergence_data = [results[s]['convergence'] for s in schedules]
plt.boxplot(convergence_data, tick_labels=schedules)
plt.ylabel('Iterations to Convergence')
plt.title('Convergence Speed')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
exploration_data = [results[s]['exploration'] for s in schedules]
plt.boxplot(exploration_data, tick_labels=schedules)
plt.ylabel('Worse Acceptance Rate')
plt.title('Exploration vs Exploitation')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

<h2> Task 5. Implement Particle Swarm Optimization algorithm </h2>
Select a multidimensional function to optimize (e.g., Rosenbrock function, Rastrigin function, or Griewank function).


1.  Define a class for the particle that includes properties such as position, velocity, personal best position, and personal best value.
2.  Initialize a swarm of particles with random positions and velocities within defined boundaries.
3.  Implement PSO formula for position update
4.  Create a running loop, which perform the optimization


```python
import numpy as np
import matplotlib.pyplot as plt


def rastrigin(x):
    A = 10
    n = len(x)
    return A * n + sum(x_i ** 2 - A * np.cos(2 * np.pi * x_i) for x_i in x)


class Particle:
    def __init__(self, dimensions, bounds):
        self.position = np.random.uniform(bounds[0], bounds[1], dimensions)
        self.velocity = np.random.uniform(-1, 1, dimensions)
        self.best_position = self.position.copy()
        self.best_value = float('inf')

    def evaluate(self, objective_func):
        current_value = objective_func(self.position)
        if current_value < self.best_value:
            self.best_value = current_value
            self.best_position = self.position.copy()
        return current_value

    def update_velocity(self, global_best_position, w, c1, c2):
        r1, r2 = np.random.random(len(self.position)), np.random.random(len(self.position))
        cognitive = c1 * r1 * (self.best_position - self.position)
        social = c2 * r2 * (global_best_position - self.position)
        self.velocity = w * self.velocity + cognitive + social

    def update_position(self, bounds):
        self.position += self.velocity
        self.position = np.clip(self.position, bounds[0], bounds[1])


def particle_swarm_optimization(objective_func, bounds, dimensions, num_particles=30,
                                max_iterations=1000, w=0.7, c1=2.0, c2=2.0):
    swarm = [Particle(dimensions, bounds) for _ in range(num_particles)]

    global_best_value = float('inf')
    global_best_position = None
    history = []

    for iteration in range(max_iterations):
        for particle in swarm:
            current_value = particle.evaluate(objective_func)

            if current_value < global_best_value:
                global_best_value = current_value
                global_best_position = particle.position.copy()

        for particle in swarm:
            particle.update_velocity(global_best_position, w, c1, c2)
            particle.update_position(bounds)

        history.append(global_best_value)

    return global_best_position, global_best_value, history


best_position, best_value, history = particle_swarm_optimization(
    rastrigin, (-5.12, 5.12), 2, num_particles=30, max_iterations=500
)

print(f"Best position: {best_position}")
print(f"Best value: {best_value:.6f}")

plt.figure(figsize=(8, 4))
plt.plot(history)
plt.xlabel('Iterations')
plt.ylabel('Best Objective Value')
plt.title('PSO Convergence')
plt.grid(True)
plt.show()
```

<h2> Task 8. Implement Ant Colony Optimization </h2>
Use the Traveling Salesman Problem (TSP) as the objective function for this task. The TSP is a classic optimization problem where the goal is to find the shortest possible route that visits each city exactly once and returns to the origin city.
<p>Objective Function: For a given set of cities and their pairwise distances, the objective function can be defined as sum of all distances or f(x)=Total Distance of the Route</p>

1.   Define the number of ants, the pheromone matrix, and the heuristic information (usually the inverse of the distance)
2.   Implement a pheromone update rule that allows pheromones to evaporate over time and be deposited by ants based on the quality of the solution found
3.   Define how ants choose their next city based on pheromone levels and heuristic information (using a probability-based approach)
4.   Finish the implementation and create loop which optimizes the objective function


```python
import numpy as np

distance_matrix = np.array([
    [0, 3.16, 4.00, 6.08, 3.00],
    [3.16, 0, 3.00, 5.10, 3.16],
    [4.00, 3.00, 0, 3.16, 1.00],
    [6.08, 5.10, 3.16, 0, 3.16],
    [3.00, 3.16, 1.00, 3.16, 0]
])

cities = ['A', 'B', 'C', 'D', 'E']


def calculate_tour_distance(tour, distance_matrix):
    distance = 0
    for i in range(len(tour)):
        distance += distance_matrix[tour[i]][tour[(i + 1) % len(tour)]]
    return distance


def construct_ant_tour(pheromones, heuristic, alpha, beta, num_cities):
    tour = [0]
    unvisited = set(range(1, num_cities))

    while unvisited:
        current_city = tour[-1]
        probabilities = []

        for city in unvisited:
            prob = (pheromones[current_city][city] ** alpha) * (heuristic[current_city][city] ** beta)
            probabilities.append(prob)

        probabilities = np.array(probabilities)
        probabilities /= probabilities.sum()

        next_city_idx = np.random.choice(len(probabilities), p=probabilities)
        next_city = list(unvisited)[next_city_idx]

        tour.append(next_city)
        unvisited.remove(next_city)

    return tour


def ant_colony_optimization(distance_matrix, num_ants=10, num_iterations=100,
                            alpha=1.0, beta=2.0, evaporation_rate=0.5, Q=1.0):
    num_cities = len(distance_matrix)
    pheromones = np.ones((num_cities, num_cities))
    heuristic = 1.0 / (distance_matrix + 1e-10)

    best_tour = None
    best_distance = float('inf')

    for iteration in range(num_iterations):
        tours = []
        distances = []

        for ant in range(num_ants):
            tour = construct_ant_tour(pheromones, heuristic, alpha, beta, num_cities)
            distance = calculate_tour_distance(tour, distance_matrix)
            tours.append(tour)
            distances.append(distance)

            if distance < best_distance:
                best_distance = distance
                best_tour = tour.copy()

        pheromones *= (1 - evaporation_rate)

        for tour, distance in zip(tours, distances):
            pheromone_deposit = Q / distance
            for i in range(len(tour)):
                from_city = tour[i]
                to_city = tour[(i + 1) % len(tour)]
                pheromones[from_city][to_city] += pheromone_deposit
                pheromones[to_city][from_city] += pheromone_deposit

    return best_tour, best_distance


best_tour, best_distance = ant_colony_optimization(distance_matrix)

print(f"Best tour: {[cities[i] for i in best_tour]}")
print(f"Best distance: {best_distance:.2f}")
```

```python
import numpy as np


def rastrigin_high_dimen(x):
    A = 10
    n = len(x)
    if n < 10:
        print("Use at least 10 dimnesions in rastrigin function")
        return -1
    else:
        return A * n + sum(x_i ** 2 - A * np.cos(2 * np.pi * x_i) for x_i in x)


def tournament_selection(population, fitness_scores):
    selected = []
    for _ in range(len(population)):
        tournament = np.random.choice(len(population), 3, replace=False)
        winner = tournament[np.argmax([fitness_scores[i] for i in tournament])]
        selected.append(population[winner].copy())
    return selected


def roulette_wheel_selection(population, fitness_scores):
    fitness_scores = np.array(fitness_scores)
    fitness_scores = fitness_scores - np.min(fitness_scores) + 1
    probabilities = fitness_scores / np.sum(fitness_scores)
    indices = np.random.choice(len(population), len(population), p=probabilities)
    return [population[i].copy() for i in indices]


def single_point_crossover(parent1, parent2):
    point = np.random.randint(1, len(parent1))
    child1 = np.concatenate([parent1[:point], parent2[point:]])
    child2 = np.concatenate([parent2[:point], parent1[point:]])
    return child1, child2


def two_point_crossover(parent1, parent2):
    points = sorted(np.random.choice(range(1, len(parent1)), 2, replace=False))
    child1 = np.concatenate([parent1[:points[0]], parent2[points[0]:points[1]], parent1[points[1]:]])
    child2 = np.concatenate([parent2[:points[0]], parent1[points[0]:points[1]], parent2[points[1]:]])
    return child1, child2


def uniform_crossover(parent1, parent2):
    mask = np.random.random(len(parent1)) < 0.5
    child1 = np.where(mask, parent1, parent2)
    child2 = np.where(mask, parent2, parent1)
    return child1, child2


def gaussian_mutation(individual, rate=0.1):
    for i in range(len(individual)):
        if np.random.random() < rate:
            individual[i] += np.random.normal(0, 0.5)
    return individual


def flip_bit_mutation(individual, rate=0.1):
    for i in range(len(individual)):
        if np.random.random() < rate:
            individual[i] = 1 - individual[i]
    return individual


population = [np.random.uniform(-5, 5, 12) for _ in range(50)]

for generation in range(100):
    fitness_scores = [-rastrigin_high_dimen(ind) for ind in population]

    selected = tournament_selection(population, fitness_scores)

    new_population = []
    for i in range(0, len(selected), 2):
        parent1, parent2 = selected[i], selected[i + 1] if i + 1 < len(selected) else selected[0]
        child1, child2 = single_point_crossover(parent1, parent2)
        new_population.extend([gaussian_mutation(child1), gaussian_mutation(child2)])

    population = new_population[:50]

best_rastrigin = population[np.argmax([-rastrigin_high_dimen(ind) for ind in population])]
print(f"Rastrigin result: {-rastrigin_high_dimen(best_rastrigin)}")

x = np.array([1.2, 2.3, 20, 3, 3.2, 5.4, 42.1, 2, 1, 2, 4, 5])
value = rastrigin_high_dimen(x)
print(f'Rastrigin function value at {x}: {value}')
```

The goal of this task is to implement basic Needleman-Wunsch algorithm for sequence alignment and to experiment with various scoring parameters to understand their impact on alignment quality.

2.   Scoring System: Define a scoring system that includes: Match Score: The score for matching characters, Mismatch Penalty: The penalty for aligning different characters, Gap Penalty: The penalty for introducing gaps in the alignment.
3.   Implement filling the Matrix: Fill in the scoring matrix according to the Needleman-Wunsch algorithm rules.
4.   Implement Traceback: After filling the matrix, perform a traceback to construct the aligned sequences.
<p>

*   S2 = "ACTGATCAA"

In order to check your implementation correctness use the following page: https://bioboot.github.io/bimm143_W20/class-material/nw/


```python
import numpy as np

def fill_in_the_matrix(seq1, seq2, match_score=2, mismatch_penalty=-1, gap_penalty=-1):
    m, n = len(seq1), len(seq2)
    matrix = np.zeros((m + 1, n + 1))

    for i in range(1, m + 1):
        matrix[i][0] = i * gap_penalty
    for j in range(1, n + 1):
        matrix[0][j] = j * gap_penalty

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if seq1[i-1] == seq2[j-1]:
                diagonal = matrix[i-1][j-1] + match_score
            else:
                diagonal = matrix[i-1][j-1] + mismatch_penalty

            up = matrix[i-1][j] + gap_penalty
            left = matrix[i][j-1] + gap_penalty

            matrix[i][j] = max(diagonal, up, left)

    return matrix

def traceback(matrix, seq1, seq2, match_score=2, mismatch_penalty=-1, gap_penalty=-1):
    i, j = len(seq1), len(seq2)
    aligned_seq1, aligned_seq2 = "", ""

    while i > 0 or j > 0:
        if i > 0 and j > 0:
            if seq1[i-1] == seq2[j-1]:
                diagonal_score = matrix[i-1][j-1] + match_score
            else:
                diagonal_score = matrix[i-1][j-1] + mismatch_penalty

            if matrix[i][j] == diagonal_score:
                aligned_seq1 = seq1[i-1] + aligned_seq1
                aligned_seq2 = seq2[j-1] + aligned_seq2
                i -= 1
                j -= 1
            elif i > 0 and matrix[i][j] == matrix[i-1][j] + gap_penalty:
                aligned_seq1 = seq1[i-1] + aligned_seq1
                aligned_seq2 = "-" + aligned_seq2
                i -= 1
            else:
                aligned_seq1 = "-" + aligned_seq1
                aligned_seq2 = seq2[j-1] + aligned_seq2
                j -= 1
        elif i > 0:
            aligned_seq1 = seq1[i-1] + aligned_seq1
            aligned_seq2 = "-" + aligned_seq2
            i -= 1
        else:
            aligned_seq1 = "-" + aligned_seq1
            aligned_seq2 = seq2[j-1] + aligned_seq2
            j -= 1

    return aligned_seq1, aligned_seq2

S1 = "ACTGACTGAACCCAA"
S2 = "ACTGATCAA"

matrix = fill_in_the_matrix(S1, S2)
aligned_s1, aligned_s2 = traceback(matrix, S1, S2)

print(f"Aligned S1: {aligned_s1}")
print(f"Aligned S2: {aligned_s2}")
print(f"Score: {matrix[-1][-1]}")
```

2.   Define a function to compute the gradient of the objective function. We will assume we do not know how to compute the gradient, therefore we have to estimate it. Implement numerical gradient estimation
3.   Update Rule: Implement the update rule for the gradient descent based on learning rate


```python
def quadratic_objective_function(x):
    return (x - 3) ** 2 + 1


def numerical_gradient(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    x_history = []
    f_history = []

    for _ in range(num_iterations):
        x_history.append(x)
        f_history.append(quadratic_objective_function(x))

        grad = numerical_gradient(quadratic_objective_function, x)
        x = x - learning_rate * grad

    return x, f_history, x_history

final_x, f_hist, x_hist = gradient_descent(5, 0.1, 20)
print(f"Final x: {final_x:.6f}")
print(f"Final f(x): {quadratic_objective_function(final_x):.6f}")
```

<p>  latest_time: The latest moment the vehicle can arrive without incurring a penalty or missing the service. This aspect is crucial for routing decisions, as it adds constraints to the schedule that must be respected. </p>

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt

customers = [
    ('Customer1', 5, (1, 2), 2, (0, 10)),
    ('Customer2', 10, (2, 3), 3, (2, 12)),
    ('Customer3', 8, (3, 1), 1, (1, 8)),
    ('Customer4', 6, (4, 4), 2, (3, 15)),
    ('Customer5', 12, (5, 5), 4, (5, 15)),
    ('Customer6', 7, (6, 2), 1, (0, 5)),
    ('Customer7', 15, (7, 8), 3, (5, 20)),
    ('Customer8', 4, (8, 3), 2, (3, 9)),
    ('Customer9', 9, (1, 7), 1, (2, 10)),
    ('Customer10', 11, (4, 6), 2, (4, 14)),
    ('Customer11', 14, (2, 5), 3, (1, 9)),
    ('Customer12', 6, (5, 7), 2, (3, 12)),
    ('Customer13', 10, (3, 3), 4, (2, 10)),
    ('Customer14', 8, (6, 5), 1, (0, 4)),
    ('Customer15', 12, (1, 4), 2, (1, 7)),
    ('Customer16', 5, (4, 2), 3, (3, 8)),
    ('Customer17', 7, (8, 6), 1, (4, 10)),
    ('Customer18', 9, (5, 3), 2, (2, 9)),
    ('Customer19', 11, (3, 6), 2, (4, 11)),
    ('Customer20', 13, (2, 8), 4, (3, 15)),
]
depot = (0, 0)
vehicles = [
    {'vehicle_id': 'Vehicle1', 'capacity': 70},
    {'vehicle_id': 'Vehicle2', 'capacity': 65},
    {'vehicle_id': 'Vehicle3', 'capacity': 80},
]
routing_rules = {
    'max_distance_per_route': 30,
    'max_customers_per_route': 5,
}

def distance(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def evaluate(solution):
    total_distance = 0
    penalty = 0
    assigned = set()

    for v_idx, route in enumerate(solution):
        if not route:
            continue

        dist = distance(depot, customers[route[0]][2])
        for i in range(len(route) - 1):
            dist += distance(customers[route[i]][2], customers[route[i+1]][2])
        dist += distance(customers[route[-1]][2], depot)
        total_distance += dist

        demand = sum(customers[i][1] for i in route)
        if demand > vehicles[v_idx]['capacity']:
            penalty += (demand - vehicles[v_idx]['capacity']) * 100
        if dist > routing_rules['max_distance_per_route']:
            penalty += (dist - routing_rules['max_distance_per_route']) * 50
        if len(route) > routing_rules['max_customers_per_route']:
            penalty += (len(route) - routing_rules['max_customers_per_route']) * 100

        assigned.update(route)

    penalty += (len(customers) - len(assigned)) * 200
    return total_distance + penalty

def initial_solution():
    indices = list(range(len(customers)))
    random.shuffle(indices)
    solution = [[] for _ in vehicles]

    for idx in indices:
        best_vehicle = min(range(len(vehicles)), key=lambda v: len(solution[v]))
        solution[best_vehicle].append(idx)

    return solution

def neighbor(solution):
    new_sol = [route[:] for route in solution]
    if random.random() < 0.5:
        routes = [i for i, r in enumerate(new_sol) if len(r) > 1]
        if routes:
            r_idx = random.choice(routes)
            i, j = random.sample(range(len(new_sol[r_idx])), 2)
            new_sol[r_idx][i], new_sol[r_idx][j] = new_sol[r_idx][j], new_sol[r_idx][i]
    else:
        from_routes = [i for i, r in enumerate(new_sol) if r]
        if len(from_routes) >= 2:
            from_idx = random.choice(from_routes)
            to_idx = random.choice([i for i in range(len(new_sol)) if i != from_idx])
            if new_sol[from_idx]:
                customer = new_sol[from_idx].pop(random.randrange(len(new_sol[from_idx])))
                new_sol[to_idx].append(customer)
    return new_sol

def simulated_annealing(temp, cooling, iterations):
    solution = initial_solution()
    best_solution = solution
    best_cost = evaluate(solution)
    costs = []

    for _ in range(iterations):
        new_solution = neighbor(solution)
        new_cost = evaluate(new_solution)
        delta = new_cost - best_cost

        if delta < 0 or random.random() < np.exp(-delta / temp):
            solution = new_solution
            if new_cost < best_cost:
                best_solution = new_solution
                best_cost = new_cost

        temp *= cooling
        costs.append(best_cost)

    return best_solution, best_cost, costs

def crossover(p1, p2):
    child = [[] for _ in vehicles]
    assigned = set()

    for i in range(len(vehicles)):
        if p1[i] and random.random() < 0.5:
            for c in p1[i][:2]:
                if c not in assigned:
                    child[i].append(c)
                    assigned.add(c)

    for c in range(len(customers)):
        if c not in assigned:
            min_route = min(range(len(vehicles)), key=lambda i: len(child[i]))
            child[min_route].append(c)

    return child

def genetic_algorithm(pop_size, mutation_rate, generations):
    population = [initial_solution() for _ in range(pop_size)]
    best_cost = float('inf')
    costs = []

    for _ in range(generations):
        scored = [(sol, evaluate(sol)) for sol in population]
        scored.sort(key=lambda x: x[1])

        if scored[0][1] < best_cost:
            best_solution = scored[0][0]
            best_cost = scored[0][1]
        costs.append(best_cost)

        new_pop = [scored[0][0]]
        while len(new_pop) < pop_size:
            p1 = random.choice(scored[:pop_size//2])[0]
            p2 = random.choice(scored[:pop_size//2])[0]
            child = crossover(p1, p2)
            if random.random() < mutation_rate:
                child = neighbor(child)
            new_pop.append(child)

        population = new_pop

    return best_solution, best_cost, costs

def hill_climber_optimize(algo_type):
    if algo_type == 'SA':
        best_params = {'temp': 100, 'cooling': 0.95, 'iter': 1000}
        for _ in range(10):
            params = best_params.copy()
            param = random.choice(['temp', 'cooling', 'iter'])
            if param == 'temp':
                params['temp'] += random.choice([-20, 20])
            elif param == 'cooling':
                params['cooling'] += random.choice([-0.02, 0.02])
            else:
                params['iter'] += random.choice([-200, 200])

            _, cost, _ = simulated_annealing(params['temp'], params['cooling'], params['iter'])
            _, best_cost, _ = simulated_annealing(best_params['temp'], best_params['cooling'], best_params['iter'])

            if cost < best_cost:
                best_params = params
        return best_params
    else:
        best_params = {'pop': 50, 'mut': 0.1, 'gen': 100}
        for _ in range(10):
            params = best_params.copy()
            param = random.choice(['pop', 'mut', 'gen'])
            if param == 'pop':
                params['pop'] += random.choice([-10, 10])
            elif param == 'mut':
                params['mut'] += random.choice([-0.05, 0.05])
            else:
                params['gen'] += random.choice([-20, 20])

            _, cost, _ = genetic_algorithm(params['pop'], params['mut'], params['gen'])
            _, best_cost, _ = genetic_algorithm(best_params['pop'], best_params['mut'], best_params['gen'])

            if cost < best_cost:
                best_params = params
        return best_params

print("1. SA - escapes local optima")
print("2. GA - natural for routing\n")

sa_params = hill_climber_optimize('SA')
ga_params = hill_climber_optimize('GA')

print("Running experiments...")
sa_times, ga_times = [], []
sa_costs, ga_costs = [], []

for _ in range(3):
    start = time.time()
    _, cost, history = simulated_annealing(sa_params['temp'], sa_params['cooling'], sa_params['iter'])
    sa_times.append(time.time() - start)
    sa_costs.append(cost)

    start = time.time()
    _, cost, history = genetic_algorithm(ga_params['pop'], ga_params['mut'], ga_params['gen'])
    ga_times.append(time.time() - start)
    ga_costs.append(cost)

print(f"\nSA: avg cost {np.mean(sa_costs):.1f}, avg time {np.mean(sa_times):.3f}s")
print(f"GA: avg cost {np.mean(ga_costs):.1f}, avg time {np.mean(ga_times):.3f}s")

_, _, sa_conv = simulated_annealing(sa_params['temp'], sa_params['cooling'], sa_params['iter'])
_, _, ga_conv = genetic_algorithm(ga_params['pop'], ga_params['mut'], ga_params['gen'])

plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.plot(sa_conv[:200], label='SA')
plt.plot(ga_conv, label='GA')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.legend()

plt.subplot(1, 2, 2)
x = [0, 1]
plt.bar(x, [np.mean(sa_costs), np.mean(ga_costs)], width=0.4)
plt.xticks(x, ['SA', 'GA'])
plt.ylabel('Average Cost')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def rastrigin(x):
    A = 10
    n = len(x)
    return A * n + sum(x_i**2 - A * np.cos(2 * np.pi * x_i) for x_i in x)

# Differential Evolution
def differential_evolution(pop_size, F, CR, max_iter, dim, bounds):
    population = np.random.uniform(bounds[0], bounds[1], (pop_size, dim))
    fitness = np.array([rastrigin(ind) for ind in population])
    best_idx = np.argmin(fitness)
    best_fitness = fitness[best_idx]
    history = []

    for _ in range(max_iter):
        for i in range(pop_size):
            a, b, c = np.random.choice(pop_size, 3, replace=False)
            mutant = population[a] + F * (population[b] - population[c])
            mutant = np.clip(mutant, bounds[0], bounds[1])

            trial = population[i].copy()
            for j in range(dim):
                if np.random.rand() < CR:
                    trial[j] = mutant[j]

            trial_fitness = rastrigin(trial)
            if trial_fitness < fitness[i]:
                population[i] = trial
                fitness[i] = trial_fitness
                if trial_fitness < best_fitness:
                    best_idx = i
                    best_fitness = trial_fitness

        history.append(best_fitness)

    return population[best_idx], best_fitness, history

# Firefly Algorithm
def firefly_algorithm(n_fireflies, alpha, beta0, gamma, max_iter, dim, bounds):
    fireflies = np.random.uniform(bounds[0], bounds[1], (n_fireflies, dim))
    brightness = np.array([rastrigin(f) for f in fireflies])
    best_idx = np.argmin(brightness)
    best_fitness = brightness[best_idx]
    history = []

    for _ in range(max_iter):
        for i in range(n_fireflies):
            for j in range(n_fireflies):
                if brightness[j] < brightness[i]:
                    r = np.linalg.norm(fireflies[i] - fireflies[j])
                    beta = beta0 * np.exp(-gamma * r**2)
                    fireflies[i] += beta * (fireflies[j] - fireflies[i]) + alpha * np.random.randn(dim)
                    fireflies[i] = np.clip(fireflies[i], bounds[0], bounds[1])
                    brightness[i] = rastrigin(fireflies[i])

                    if brightness[i] < best_fitness:
                        best_idx = i
                        best_fitness = brightness[i]

        alpha *= 0.95
        history.append(best_fitness)

    return fireflies[best_idx], best_fitness, history

# Time complexity analysis
print("Time Complexity Analysis:")
print("=" * 60)

dimensions = [2, 5, 10, 15, 20]
de_times = []
ff_times = []

for dim in dimensions:
    start = time.time()
    differential_evolution(30, 0.8, 0.9, 100, dim, (-5.12, 5.12))
    de_times.append(time.time() - start)

    start = time.time()
    firefly_algorithm(30, 0.5, 1.0, 0.01, 100, dim, (-5.12, 5.12))
    ff_times.append(time.time() - start)

print(f"DE: O(pop_size × max_iter × dim)")
print(f"Firefly: O(n_fireflies² × max_iter × dim)")

# Growth rate analysis
print("\nGrowth Rate Analysis:")
for i in range(1, len(dimensions)):
    de_growth = (de_times[i] - de_times[i-1]) / (dimensions[i] - dimensions[i-1])
    ff_growth = (ff_times[i] - ff_times[i-1]) / (dimensions[i] - dimensions[i-1])
    print(f"Dim {dimensions[i-1]}→{dimensions[i]}: DE={de_growth:.4f}s/dim, FF={ff_growth:.4f}s/dim")

# Hyperparameter experiments
print("\nHyperparameter Optimization:")
print("=" * 60)

dim = 10
bounds = (-5.12, 5.12)

# DE experiments
F_values = [0.3, 0.5, 0.8, 1.0]
CR_values = [0.1, 0.5, 0.7, 0.9]
de_results = {}

for F in F_values:
    for CR in CR_values:
        costs = []
        for _ in range(5):
            _, cost, _ = differential_evolution(30, F, CR, 200, dim, bounds)
            costs.append(cost)
        de_results[(F, CR)] = np.mean(costs)

# Firefly experiments
alpha_values = [0.1, 0.5, 1.0]
gamma_values = [0.001, 0.01, 0.1]
ff_results = {}

for alpha in alpha_values:
    for gamma in gamma_values:
        costs = []
        for _ in range(5):
            _, cost, _ = firefly_algorithm(30, alpha, 1.0, gamma, 200, dim, bounds)
            costs.append(cost)
        ff_results[(alpha, gamma)] = np.mean(costs)

best_de = min(de_results, key=de_results.get)
best_ff = min(ff_results, key=ff_results.get)
print(f"Best DE params: F={best_de[0]}, CR={best_de[1]}, cost={de_results[best_de]:.2f}")
print(f"Best Firefly params: alpha={best_ff[0]}, gamma={best_ff[1]}, cost={ff_results[best_ff]:.2f}")

# Run with best parameters
_, _, de_hist = differential_evolution(50, best_de[0], best_de[1], 500, 10, bounds)
_, _, ff_hist = firefly_algorithm(50, best_ff[0], 1.0, best_ff[1], 500, 10, bounds)

# Visualization
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(dimensions, de_times, 'o-', label='DE')
plt.plot(dimensions, ff_times, 's-', label='Firefly')
plt.xlabel('Dimensions')
plt.ylabel('Time (s)')
plt.title('Time Complexity')
plt.legend()

plt.subplot(1, 3, 2)
de_matrix = np.zeros((4, 4))
for i, F in enumerate([0.3, 0.5, 0.8, 1.0]):
    for j, CR in enumerate([0.1, 0.5, 0.7, 0.9]):
        de_matrix[i, j] = de_results[(F, CR)]
plt.imshow(de_matrix, cmap='viridis', aspect='auto')
plt.colorbar(label='Cost')
plt.xlabel('CR')
plt.ylabel('F')
plt.title('DE Hyperparameters')
plt.xticks([0,1,2,3], ['0.1','0.5','0.7','0.9'])
plt.yticks([0,1,2,3], ['0.3','0.5','0.8','1.0'])

plt.subplot(1, 3, 3)
plt.plot(de_hist[:200], label='DE')
plt.plot(ff_hist[:200], label='Firefly')
plt.xlabel('Iteration')
plt.ylabel('Best Cost')
plt.title('Convergence')
plt.legend()

plt.tight_layout()
plt.show()

# Non-convergence analysis
print("\nNon-convergence Analysis:")
print("=" * 60)

# DE with F=0
_, cost_f0, hist_f0 = differential_evolution(30, 0.0, 0.9, 100, 10, bounds)
# DE with CR=0
_, cost_cr0, hist_cr0 = differential_evolution(30, 0.8, 0.0, 100, 10, bounds)
# Firefly with large gamma
_, cost_gamma_large, hist_gamma_large = firefly_algorithm(30, 0.5, 1.0, 10.0, 100, 10, bounds)

plt.figure(figsize=(8, 4))
plt.plot(hist_f0, label='DE (F=0)')
plt.plot(hist_cr0, label='DE (CR=0)')
plt.plot(hist_gamma_large, label='Firefly (γ=10)')
plt.xlabel('Iteration')
plt.ylabel('Best Cost')
plt.title('Non-convergence Examples')
plt.legend()
plt.show()

print("DE with F=0: No mutation → no exploration → stuck at initial random solution")
print("DE with CR=0: No crossover → no information exchange → poor convergence")
print("Firefly with large γ: Attraction decays too fast → fireflies don't move → no optimization")
```
